```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
unit = pd.read_csv('/content/drive/MyDrive/BPDV65/opendata_unittype.csv')
unit
```

```python
eat = pd.read_csv('/content/drive/MyDrive/BPDV65/opendata_eating_score.csv')
eat
```

```python
unit1=unit.dropna(subset=['propertytype_name_th','price_min','area_usable_min'])
```

```python
eat1=eat.dropna(subset=['province_name_th'])
```

```python
data1 = unit1.merge(eat1, left_on='row_number', right_on='row_number')
data1.head()
```

```python
data2 = data1[data1['province_name_th']=='เชียงใหม่']
data2
```

```python
home =  data2[data2['propertytype_name_th']=='บ้าน']
home
```

```python
condo =  data2[data2['propertytype_name_th']=='คอนโด']
condo
```

```python
compare = pd.concat([home,condo])
compare
```

```python
set(compare['propertytype_name_en'])
```

```python
compare['propertytype_name_en'].unique()
```

```python
pd.plotting.parallel_coordinates(compare, 'propertytype_name_en',cols = ['price_min','area_usable_min'], color=['#DD0000', '#33CC00'])
```
